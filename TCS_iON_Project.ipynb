### Feature Engineering

```python
df_train.columns
```

```python
#combing sc_w, sc_h, m_dep to a single size column
df_train['mob_size']=df_train['sc_h']*df_train['sc_w']*df_train['m_dep']

#dropping the screen height and width columns
df_train=df_train.drop(['sc_h','sc_w','m_dep'],axis=1)
```

```python
#dropping the same column in test
#dropping the id column in test data
df_test=df_test.drop(['id'],axis=1)

#we have screen width and screen height,we will make create a new column as mobile size with those two
df_test['mob_size']=df_test['sc_h']*df_test['sc_w']*df_test['m_dep']

#dropping the screen height and width columns
df_test=df_test.drop(['sc_h','sc_w','m_dep'],axis=1)
```

### Split the data into X and y


```python
X=df_train.drop(['price_range'],axis=1)
y=df_train['price_range']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_test2 = df_test.copy()
```

### Modelling without scaling

```python
X.head()
```

```python
#Fitting with different models
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier , GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models={'RandomForestClassifier':RandomForestClassifier(n_estimators=300),
        'ExtraTreesClassifier':ExtraTreesClassifier(),
        'GradientBoostingClassifier':GradientBoostingClassifier(),
        'LogisticRegression':LogisticRegression(),
        'KNeighborsClassifier':KNeighborsClassifier(),
        'SVC':SVC(kernel='rbf'),
        'DecisionTreeClassifier':DecisionTreeClassifier(criterion='entropy'),
        'GaussianNB':GaussianNB(),
       }

for name,model in models.items():
    model.fit(x_train,y_train)
    print(name+' trained')
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix
model_test_f1_score=[]
model_test_accuracy_score=[]
model_confusion_matrix=[]

#checking the scores of classifiers
for name,model in models.items():
    y_pred=model.predict(x_test)
    model_test_f1_score.append(np.round(f1_score(y_test,y_pred,average='weighted')*100,2))
    model_test_accuracy_score.append(np.round(accuracy_score(y_test,y_pred)*100,2))
    model_confusion_matrix.append(confusion_matrix(y_test,y_pred))
model_confusion_matrix
```

```python
Models = ['Random Forest','ExtraTreesClassifier','GradientBoostingClassifier','Logistic regression','KNN','SVM','DecisionTree','GaussianNB']
model_performance = pd.DataFrame({'Model':Models,
   'F1_score':model_test_f1_score,
   'Accuracy_Score':model_test_accuracy_score
   })
model_performance
```

```python
px.histogram(data_frame=model_performance,x='Model',y='Accuracy_Score',title='Comparison of accuracies of different model')
```

### Modelling after scaling the data


```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
X=pd.DataFrame(X)
X_test2=scaler.fit_transform(X_test2)

x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
models={'RandomForestClassifier':RandomForestClassifier(n_estimators=300),
        'ExtraTreesClassifier':ExtraTreesClassifier(),
        'GradientBoostingClassifier':GradientBoostingClassifier(),
        'LogisticRegression':LogisticRegression(),
        'KNeighborsClassifier':KNeighborsClassifier(),
        'SVC':SVC(kernel='rbf'),
        'DecisionTreeClassifier':DecisionTreeClassifier(criterion='entropy'),
        'GaussianNB':GaussianNB(),
       }

for name,model in models.items():
    model.fit(x_train,y_train)
    print(name+' trained')
```

```python
from sklearn.metrics import f1_score,accuracy_score,confusion_matrix
model_test_f1_score=[]
model_test_accuracy_score=[]
model_confusion_matrix=[]

#checking the scores of classifiers
for name,model in models.items():
    y_pred=model.predict(x_test)
    model_test_f1_score.append(np.round(f1_score(y_test,y_pred,average='weighted')*100,2))
    model_test_accuracy_score.append(np.round(accuracy_score(y_test,y_pred)*100,2))
    model_confusion_matrix.append(confusion_matrix(y_test,y_pred))
model_confusion_matrix
```

```python
Models = ['Random Forest','ExtraTreesClassifier','GradientBoostingClassifier','Logistic regression','KNN','SVM','DecisionTree','GaussianNB']
model_performance = pd.DataFrame({'Model':Models,
   'F1_score':model_test_f1_score,
   'Accuracy_Score':model_test_accuracy_score
   })
model_performance
```

Logistic regression gave the best model score after standardization. Hence selecting Logistic regression as the best model
